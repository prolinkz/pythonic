#### Beginner’s Python Cheat Sheet - If Statements and While Loops

* If statements allow you to execute a block of code if a certain condition is true.
* While loops allow you to repeat a block of code as long as a particular condition remains true.
- Here's an example:

### If statement

<code>
x = 5
if x > 0:
print("x is greater than zero")
else:
print("x is not greater than zero")
###### Output: "x is greater than zero"
</code>

### While loop
<code>
count = 1
while count <= 5:
print(count)
count += 1
###### Output: 1, 2, 3, 4, 5
</code>
- Note that the indentation in Python is crucial for correct syntax. Also note that while loops can be infinite if the condition never becomes false! Always
- Note that the indentation in Python is very important for correct syntax. The first line after each colon must be indented by four spaces or one tab
